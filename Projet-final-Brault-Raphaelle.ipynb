## Analyses statistiques

###### Tests-t pour deux échantillons

Nous utiliserons un p-value de 0.05 pour les analyses

```python
#Importation du module scipy pour les tests-t

import scipy
scipy.__version__

from scipy import stats
```

```python
#Test-t pour la longueur des sépales

setosa_longsepales = iris[iris['Species'] == 0]['Sepal.Length']
versicolor_longsepales = iris[iris['Species'] == 1]['Sepal.Length']

stats.ttest_ind(setosa_longsepales, versicolor_longsepales)

#Différence significative (t(43) = -8.017, p < 0.05) entre les deux groupes pour la longueur des sépales
```

```python
#Test-t pour la largeur des sépales

setosa_largsepales = iris[iris['Species'] == 0]['Sepal.Width']
versicolor_largsepales = iris[iris['Species'] == 1]['Sepal.Width']

stats.ttest_ind(setosa_largsepales, versicolor_largsepales)

#Différence significative (t(43) = 2.932, p < 0.05) entre les deux groupes pour la largeur des sépales
```

```python
#Test-t pour la longueur des pétales

setosa_longpetales = iris[iris['Species'] == 0]['Petal.Length']
versicolor_longpetales = iris[iris['Species'] == 1]['Petal.Length']

stats.ttest_ind(setosa_longpetales, versicolor_longpetales)

#Différence significative (t(43) = -25.653, p < 0.05) entre les deux groupes pour la longueur des pétales
```

```python
#Test-t pour la largeur des pétales

setosa_largpetales = iris[iris['Species'] == 0]['Petal.Width']
versicolor_largpetales = iris[iris['Species'] == 1]['Petal.Width']

stats.ttest_ind(setosa_largpetales, versicolor_largpetales)

#Différence significative (t(43) = -16.583, p < 0.05) entre les deux groupes pour la largeur des pétales
```

###### Etape 1: Calcul des vecteurs moyens de dimension d

```python
mean_vectors = []
for cl in range(1,4):
    mean_vectors.append(np.mean(x[y==cl], axis=0))
    print('Mean Vector class {0}: {1}\n'.format(cl, mean_vectors[cl-1]))
```

###### Étape 2: Calcul des matrices de dispersion

```python
S_W = np.zeros((4,4))
for cl,mv in zip(range(1,4), mean_vectors):
    class_sc_mat = np.zeros((4,4))                 
    for row in x[y == cl]:
        row, mv = row.reshape(4,1), mv.reshape(4,1) 
        class_sc_mat += (row-mv).dot((row-mv).T)
    S_W += class_sc_mat                             
print('within-class Scatter Matrix:\n', S_W)


# Matrice de dispersion entre les classes

overall_mean = np.mean(x, axis=0)

S_B = np.zeros((4,4))
for i,mean_vec in enumerate(mean_vectors):  
    n = x[y==i+1,:].shape[0]
    mean_vec = mean_vec.reshape(4,1) # make column vector
    overall_mean = overall_mean.reshape(4,1) # make column vector
    S_B += n * (mean_vec - overall_mean).dot((mean_vec - overall_mean).T)

print('between-class Scatter Matrix:\n', S_B)
```

###### Étape 3: Résoudre le problème généralisé des valeurs propres

```python
eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))

for i in range(len(eig_vals)):
    eigvec_sc = eig_vecs[:,i].reshape(4,1)   
    print('\nEigenvector {}: \n{}'.format(i+1, eigvec_sc.real))
    print('Eigenvalue {:}: {:.2e}'.format(i+1, eig_vals[i].real))
```

###### Étape 4: Sélection des discriminants linéaires pour le nouvel sous-espace de fonctionnalités

```python
# Faites une liste de tuples (valeur propre, vecteur propre)

eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Trier les tuples (valeur propre, vecteur propre) de haut en bas

eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)

# Confirmer visuellement que la liste est correctement triée par diminution des valeurs propres

print('Eigenvalues in decreasing order:\n')
for i in eig_pairs:
    print(i[0])
```

```python
print('Variance explained:\n')
eigv_sum = sum(eig_vals)
for i,j in enumerate(eig_pairs):
    print('eigenvalue {0:}: {1:.2%}'.format(i+1, (j[0]/eigv_sum).real))
```

```python
#Le premier couple explique 98,66% de la variance
```

```python
W = np.hstack((eig_pairs[0][1].reshape(4,1), eig_pairs[1][1].reshape(4,1)))
print('Matrix W:\n', W.real)
```

###### Étape 5: Transformer les échantillons dans le nouveau sous-espace

```python
X_lda = x.dot(W)
```

```python
def plot_lda_by_step():

    ax = plt.subplot(111)
    for label,marker,color in zip(
        range(1,3),('^', 's'),('blue', 'red')):

        plt.scatter(x=X_lda[:,0].real[y == label],
                y=X_lda[:,1].real[y == label],
                marker=marker,
                color=color,
                alpha=0.5,
                label=label_dict[label]
                )

    plt.xlabel('LD1')
    plt.ylabel('LD2')

    leg = plt.legend(loc='upper right', fancybox=True)
    leg.get_frame().set_alpha(0.5)
    plt.title('LDA: Iris projection onto the first 2 linear discriminants')

    # hide axis ticks
    plt.tick_params(axis="both", which="both", bottom="off", top="off",  
            labelbottom="on", left="off", right="off", labelleft="on")

    # remove axis spines
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)    

    plt.grid()
    plt.tight_layout
    plt.show()

plot_lda_by_step()
```

## Apprentissage non-supervisé (ACP)

###### Objectif

Demander s’il existe une représentation correspondante de la dimension inférieure conservant les principales caractéristiques des données, soit la longueur des sépales, la largeur des sépales, la longueur des pétales et la largeur des pétales. La réduction de dimensionnalité avec ACP implique la mise à zéro d'un ou de plusieurs des composants principaux les plus petits. Le résultat est une projection dimensionnelle inférieure des données, mais conserve la variance maximale des données.


```python
#Importation de sklearn, warnings et matplotlib inline
import sklearn

import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter("ignore")
print('sklearn version:', sklearn.__version__)

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
X_iris = iris.drop('Species', axis=1)
y_iris = iris['Species']


# Étape 1: définir l'algorithme

from sklearn.decomposition import PCA

# Étape 2 = définir les hyperparamètres pour chaque algorithme

model = PCA(n_components=2)

# Étape 3 = adaptation

model.fit(X_iris)

# étape 4 = transformation

X_2D = model.transform(X_iris)

iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]

# On recoit ainsi 2 composantes
```

```python
#Le graphique actuel montre qu'une représentation en 2D a une bonne séparation entre setosa et versicolor

sns.lmplot(x = "PCA1", y = "PCA2", hue='Species', data=iris, fit_reg=False);

#(Ne pas tenir compte de virginica)
```
